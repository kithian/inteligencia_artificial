## Variable elegida por el equipo que será analizada

✈ La variable elegida por el equipo fue la de  coffee_name

`````python
top_coffee_name = data['coffee_name'].value_counts().nlargest(5)
selected_data = data[data['coffee_name'].isin(top_coffee_name.index)]

plt.figure(figsize=(12, 8))
plt.title('Distribución de Monto por Tipo de Café (Top 5)', fontsize=16, fontweight="bold")
sns.violinplot(data=selected_data, x='coffee_name', y='money', palette='Set2')
plt.xlabel('Tipo de Café', fontsize=14, fontweight="bold")
plt.ylabel('Monto ($)', fontsize=14, fontweight="bold")
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
`````

---
✅ **Interpretación**

---
Los violines que son más delgados dan a entender que mantienen un estándar en sus precios; en cambio, los violines más anchos pueden ser porque tienen diferentes tamaños de vaso o puede que sean personalizados.

Los violines que se encuentran más arriba pueden dar a entender que son de mayor calidad 'premium' y, por ende, más caros; a su vez, los que están más abajo son cafés más económicos.

Las variables "Weekdaysort" y "Monthsort" muestran una correlación
