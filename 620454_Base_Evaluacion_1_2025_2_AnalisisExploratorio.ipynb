`````python
# Variable continua: money
print("4. MEDIDAS DE VARIABLE CONTINUA (money):")
medidas_money = {
    'Media': data['money'].mean(),
    'Mediana': data['money'].median(),
    'Moda': data['money'].mode()[0],
    'Desviación estándar': data['money'].std(),
    'Mínimo': data['money'].min(),
    'Máximo': data['money'].max(),
    'Rango': data['money'].max() - data['money'].min(),
    'Q1': data['money'].quantile(0.25),
    'Q3': data['money'].quantile(0.75)
}

for medida, valor in medidas_money.items():
    print(f"{medida}: {valor:.2f}")

# 5. Interpretación medidas variable continua
print("\n5. INTERPRETACIÓN:")
print("""
- La media de las ventas es de aproximadamente $36.56, indicando el valor promedio por transacción
- La mediana ($38.70) es mayor que la media, sugiriendo una ligera asimetría negativa
- La moda es $38.70, siendo el precio más frecuente en las transacciones
- La desviación estándar de $9.16 indica variabilidad moderada en los precios
- El rango de $31.60 muestra diferencia significativa entre precios mínimos y máximos
- El 25% de las transacciones son menores a $28.90 (Q1) y el 75% menores a $38.70 (Q3)
""")

# 6. Gráfico de variable continua
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(data['money'], kde=True, bins=30)
plt.title('Distribución de Montos de Venta')
plt.xlabel('Monto ($)')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
sns.boxplot(y=data['money'])
plt.title('Diagrama de Caja - Montos de Venta')
plt.ylabel('Monto ($)')

plt.tight_layout()
plt.show()

# 7. Dos comentarios acerca de la interpretación gráfica
print("\n7. COMENTARIOS GRÁFICOS:")
print("""
1. La distribución de los montos presenta múltiples picos, sugiriendo que existen
   precios específicos y comunes para diferentes tipos de café
2. El diagrama de caja muestra varios valores atípicos en el rango inferior, indicando
   que hay transacciones con precios significativamente más bajos que la mayoría
""")
`````
