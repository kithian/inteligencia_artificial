## Análisis univariado

### Variable continua

Deberá incluir el cálculo e interpretación de las siguientes medidas estadísticas:
+ Medidas de posición (primer, segundo y tercer cuartil)
+ Desviación estándar
+ Promedio

`````python
# Variable continua: money
medidas_money = {
    'Media': data['money'].mean(),
    'Mediana': data['money'].median(),
    'Moda': data['money'].mode()[0],
    'Desviación estándar': data['money'].std(),
    'Mínimo': data['money'].min(),
    'Máximo': data['money'].max(),
    'Rango': data['money'].max() - data['money'].min(),
    'Q1': data['money'].quantile(0.25),
    'Q3': data['money'].quantile(0.75)
}

for medida, valor in medidas_money.items():
    print(f"{medida}: {valor:.2f}")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(data['money'], kde=True, bins=30)
plt.title('Distribución de Montos de Venta')
plt.xlabel('Monto ($)')
plt.ylabel('Frecuencia')

plt.subplot(1, 2, 2)
sns.boxplot(y=data['money'])
plt.title('Diagrama de Caja - Montos de Venta')
plt.ylabel('Monto ($)')

plt.tight_layout()
plt.show()
`````

---
✅ **Interpretación**

---

La media de las ventas es de aproximadamente 36.56, indicando el valor promedio por transacción

La mediana (38.70) es mayor que la media, sugiriendo una ligera asimetría negativa

La moda es 38.70, siendo el precio más frecuente en las transacciones
La desviación estándar de 9.16 indica variabilidad moderada en los precios

El rango de 31.60 muestra diferencia significativa entre precios mínimos y máximos

El 25% de las transacciones son menores a 28.90 (Q1) y el 75% menores a 38.70 (Q3)


Gráfico:

La distribución de los montos presenta múltiples picos, sugiriendo que existen precios específicos y comunes para diferentes tipos de café

El diagrama de caja muestra varios valores atípicos en el rango inferior, indicando que hay transacciones con precios significativamente más bajos que la mayoría

# Análisis Bivariado

En esta sección deberá elegir un par de variables que sean de interes de analizar e interpretar en el negocio los resultados obtenidos

`````python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='coffee_name', y='hour_of_day')
plt.title('Distribución de Horas de Venta por Tipo de Café')
plt.xlabel('Tipo de Café')
plt.ylabel('Hora del Día')
plt.xticks(rotation=45)
# Mostrar todas las horas del 6 al 23
plt.yticks(range(6, 24, 1))  # el '1' es el paso

plt.show()
`````

Respecto de los tipos de café más vendidos y las horas del día:

---Tendencia central (mediana de horas de venta)

Las medianas se concentran entre las 12:00 y 17:00 horas, indicando que el mayor consumo ocurre entre mediodía y tarde.

Hot Chocolate y Cocoa presentan las medianas más altas (16:00–17:00), con mayor preferencia vespertina.

Americano y Cortado tienen las medianas más bajas (~12:00–13:00), lo que refleja consumo más temprano.

---Dispersión (caja e IQR)

Latte, Espresso y Cappuccino muestran rangos amplios, con ventas distribuidas a lo largo de todo el día.

Hot Chocolate y Cocoa tienen cajas más concentradas hacia la tarde, con un patrón más definido.

---Outliers

No se observan valores atípicos fuertes; las ventas se distribuyen de forma continua entre las 6–7 a.m. y las 10 p.m.

Comparación general

Consumo matutino: Americano, Cortado y Americano with Milk.

Consumo vespertino: Hot Chocolate, Cocoa y Cappuccino.

Consumo equilibrado en todo el día: Latte y Espresso.

✅ Interpretación con sentido de negocio:

El gráfico revela que, aunque todos los tipos de café se venden en un rango amplio de horas, existen patrones definidos: los cafés Americano, Cortado y Americano with Milk se prefieren en la mañana, mientras que Hot Chocolate, Cocoa y Cappuccino. Para el negocio, esto sugiere diseñar promociones y disponibilidad ajustadas a la hora del día para maximizar ventas y satisfacer hábitos de consumo.

# Análisis de simetría/asimetría

## Variable elegida por el equipo que será analizada

✈ La variable elegida por el equipo fue la de  coffee_name

`````python
top_coffee_name = data['coffee_name'].value_counts().nlargest(5)
selected_data = data[data['coffee_name'].isin(top_coffee_name.index)]

plt.figure(figsize=(12, 8))
plt.title('Distribución de Monto por Tipo de Café (Top 5)', fontsize=16, fontweight="bold")
sns.violinplot(data=selected_data, x='coffee_name', y='money', palette='Set2')
plt.xlabel('Tipo de Café', fontsize=14, fontweight="bold")
plt.ylabel('Monto ($)', fontsize=14, fontweight="bold")
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
`````

---
✅ **Interpretación**

---
Los violines que son más delgados dan a entender que mantienen un estándar en sus precios; en cambio, los violines más anchos pueden ser porque tienen diferentes tamaños de vaso o puede que sean personalizados.

Los violines que se encuentran más arriba pueden dar a entender que son de mayor calidad 'premium' y, por ende, más caros; a su vez, los que están más abajo son cafés más económicos.
